```python
import numpy as np
```

```python
import matplotlib.pyplot as plt # It imports the core plotting library in Python, enabling you to create various types of visualizations like charts and graphs.
```

```python
from PIL import Image # PIL (Python Imaging Library) provides image processing capabilities within Python.
                        # Image should should start with capital i;e., "I"
```

```python
lion_img = Image.open(r'C:\Users\HP\Downloads\lion.jpg') # 'r' is a raw string literal.
lion_img                                              # And to read the image in definitely we need to use 'r'
```

```python
type(lion_img) # It gives the image type
```

```python
lion_arr = np.asarray(lion_img) # converting image into an array
lion_arr
```

```python
type(lion_arr) # the image is a nd array.
```

```python
plt.imshow(lion_img) # plt.imshow() takes the image data and renders it as a visual image within a Matplotlib figure.
```

```python
lion_arr.shape # The output will come as tuple. It will typically have three values: (height, width, channels).
                # height: The number of pixels in the vertical direction.
                # width: The number of pixels in the horizontal direction.
                # channels: The number of color channels (usually 3 for RGB: red, green, blue).
```

```python
lion_arr1 = lion_arr.copy() # copying
lion_arr1
```

```python
lion_arr == lion_arr1 # checking whether the two arrays are true or false
```

```python
plt.imshow(lion_arr1[:,:,0]) # [:,:,0]: This is NumPy array slicing. Let's break it down:
                            # (first colon): Selects all rows of the array.
                            # (second colon): Selects all columns of the array.
                            # Selects the element at index 0 along the third dimension. 
                            # In an RGB image, the third dimension represents the color channels: 0 for red, 1 for green, and 2 for blue.
```
